```python
from IPython.display import IFrame
import base64, os, io
```

```python
# MAKE BASE64 EMBED OF A PDF FILE
```

```python
# IPython.display.IFrame CAN ONLY READ IN AND VIEW A PDF FILE FROM LOCAL JUPYTER SESSION PATH, SO MUST 
# DECODE base64 EMBEDDING, SAVE TO THAT PATH (THE JUPYTER SESSION'S DEFAULT PATH), AND OPEN THE PERSISTED PDF.
```

```python
# ENCODE
def pdf_encoder(path_and_filename_of_pdf):

    filename = 'base64_pdf_dump.txt'

    if os.name=='nt': desktop_path = r'c:\users\bill\desktop'
    elif os.name=='posix': desktop_path = r'/home/bear/Desktop'

    base64_path_and_file_name = os.path.join(desktop_path, filename)
    
    with open(base64_path_and_file_name, 'w') as write_file:
        with open(rf'{path_and_filename_of_pdf}', 'rb') as read_file:
            write_file.write(base64.b64encode(read_file.read()).decode("utf-8"))
```
