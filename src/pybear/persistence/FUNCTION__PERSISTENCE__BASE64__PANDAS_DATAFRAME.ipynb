```python
import numpy as np
import pandas as pd
import scipy.sparse as ss
import os, io, base64
```

```python
# ENCODE

def pandas_df_encoder(PD_DF):
    
    if os.name=='nt': desktop_path = r'c:\users\bill\desktop'
    elif os.name=='posix': desktop_path = r'/home/bear/Desktop'

    filename = 'base64_pd_df_dump.txt'
    
    path_and_file_name = os.path.join(desktop_path, filename)
    
    # ENCODE DATAFRAME ####################################################################################
    with open(path_and_file_name, 'w') as f:
        ENCODED_DF = PD_DF.to_csv(index=False)
        f.write(base64.b64encode(ENCODED_DF.encode('utf-8')).decode('utf-8'))

    del ENCODED_DF
    # END ENCODE DATAFRAME #################################################################################

```

```python
# CREATE DUMMY DATAFRAME
DATA = np.random.randint(0,10,(20_000,10),dtype=np.uint8)
HEADER = np.fromiter(map(lambda x: str(x).rjust(2, '0'), np.linspace(1, 10, num=10, dtype=np.uint16).astype('<U2')), dtype='<U2')

PD_DF = pd.DataFrame(data=DATA, columns=HEADER, dtype=np.uint8)

del DATA, HEADER
```

```python
# ENCODE
pandas_df_encoder(PD_DF)
```
